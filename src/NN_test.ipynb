### 4)只选取protein数据，以作简单分类

```python
protein_char = pdb_csv[pdb_csv.macromoleculeType == 'Protein']
protein_seq = pdb_seq_csv[pdb_seq_csv.macromoleculeType == 'Protein']
alldata=pd.merge(protein_char,protein_seq,on='structureId')
alldata.index=alldata['structureId']
alldata=alldata[alldata['sequence'].notna()]
alldata.head()
```

### 5)通过自己的编程实现特征构建

>大致说明，就是将蛋白的多肽序列转化为矩阵

**一个例子如下：**

![features](images/features.png) `_A`表示以`A`开头，`A_`表示以`A`结尾

**只提取分类信息**

```python
uniqId = alldata['structureId'].unique()
j=0
dataset = []
for i in uniqId:
    j = j+1
    if j<100:
        mypattern = alldata.loc[i,'sequence']
        if type(mypattern) == str:
            alldata.loc[i,'features'] = dataset.append([alldata.loc[i,'classification'],get_amino_acid_pattern(mypattern)])
        else:
            alldata.loc[i,'features'] = dataset.append([alldata.loc[i,'classification'].iloc[0],get_amino_acid_pattern(*mypattern)])
dataset = pd.DataFrame(dataset,columns=['class','features'])
dataset.head()
```

### 6)从数据集中展示几个例子图

```python
picset = dataset.sample(10)
plt.figure(1,figsize=(10,25), dpi=300)
for i in range(10):
    ax = plt.subplot2grid((5, 2), (i%5,i%2 ), colspan = 1, rowspan = 1)
    ax.set_title(picset.iloc[i,0])
    ax.imshow(picset.iloc[i,1],cmap=plt.cm.RdBu)
```

**图中的小点（不是屏幕上的灰尘）就是关键点。但是目前人眼不太能分别出特征，所以准备构建一个CNN来进行学习**
